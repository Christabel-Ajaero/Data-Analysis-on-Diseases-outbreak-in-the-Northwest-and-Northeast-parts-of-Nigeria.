## CLEANING THE DATA

```python
diseas.sort_values('state')
```

```python
dis = diseas['state'].value_counts()
```

```python
sorted_d = diseas.sort_values('state')
```

```python
state_sort = sorted_d.set_index('state')
```

```python
state_sort.drop('serotype', axis = 1)
```

```python
listing = ['Abia','Akwa Ibom','Bayelsa', 'Cross River', 'Delta', 'Ebonyi', 'Edo','Ekiti', 'Enugu', 
          'Lagos', 'Ogun', 'Ondo', 'Osun', 'Oyo', 'Rivers', 'Plateau','Kwara', 'Nasarawa', 'Kogi', 'Niger',
          'Benue', 'Federal Capital Territory','Imo', 'Anambra']
new_sorted= state_sort.drop(listing, axis= 0)
new_sorted
```

```python
del_sero = new_sorted.drop('serotype', axis = 1)
del_sero
```

```python


del_disease = del_sero[del_sero.disease != 'Ebola']
del_disease1 = del_disease[del_disease.disease != 'Malaria']
del_disease2 = del_disease1[del_disease1.disease != 'Marburg Virus']
del_disease3 = del_disease2[del_disease2.disease != 'Viral Haemmorrhaphic Fever']
del_disease3
```

```python
del_disease3['disease'].unique()
```

```python
mytable = del_disease3.drop('id', axis = 1)
```

```python

next_table = mytable.reset_index()
```

```python
final_table = next_table.replace(['Rubella Mars', 'Yellow Fever'], ['Rubella_Mars', 'Yellow_Fever'])

Cholera = final_table[final_table.disease == 'Cholera'] 


Measles = final_table[final_table.disease == 'Measles']

Rubella_Mars = final_table[final_table.disease == 'Rubella_Mars']

Meningitis = final_table[final_table.disease == 'Meningitis']

Diarrhoea = final_table[final_table.disease == 'Diarrhoea']

Yellow_Fever = final_table[final_table.disease == 'Yellow_Fever']

final_table
```

## VISUALIZATION

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
b = sns.countplot(x = 'disease', hue = 'gender', data = final_table, palette = 'rainbow')
b.set_xticklabels(b.get_xticklabels(), rotation= 20)
plt.title('disease against gender')
plt.legend(loc='lower left')
plt.show()
```
